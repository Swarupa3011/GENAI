```python
import torch
from diffusers import PixArtAlphaPipeline
```

```python
model_id="PixArt-alpha/PixArt-XL-2-512x512"
dtype = torch.float16 if torch.cuda.is_available() else torch.float32

pipe = PixArtAlphaPipeline.from_pretrained(
    model_id,
    torch_dtype=dtype,
)

device = "cuda" if torch.cuda.is_available() else "cpu"

pipe = pipe.to(device)

#Optional: enable faster attention if available
if device == "cuda":
    pipe.enable_model_cpu_offload() # works well on limited VRAM
    pipe.enable_vae_slicing()
```

```python
# 2) Write a good prompt for a girl's face
prompt = (
    "Ultra-realistic close-up portrait of a young woman, soft smile, "
    "long wavy brown hair, brown eyes, smooth skin, natural makeup, "
    "shot with 85mm lens, cinematic lighting, high detail skin, photographic quality"
)

negative_prompt = (
    "blurry, low quality, extra fingers, extra eyes, deformed, distorted, "
    "asymmetrical face, bad anatomy, text, watermark, logo"
)
```

```python
generator = torch.Generator(device=device).manual_seed(42) # set seed for reproducibility

image = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=24, # increase for slightly higher quality
    guidance_scale=4.5, #3.5-6 is typical
    height=512,
    width=512,
    generator=generator
).images[0]

#4) Save the result

image.save("face.png")

print("Saved to face.png")
```
